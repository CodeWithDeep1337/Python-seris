## List methods 

```python
# ===============================
# Python List Methods Demo
# ===============================

# Initial list
numbers = [10, 20, 30, 40]
print("Initial list:", numbers)

# 1. append() – Add element at end
numbers.append(50)
print("append(50):", numbers)

# 2. extend() – Add multiple elements
numbers.extend([60, 70])
print("extend([60, 70]):", numbers)

# 3. insert() – Insert at specific index
numbers.insert(2, 25)
print("insert(2, 25):", numbers)

# 4. remove() – Remove first occurrence
numbers.remove(30)
print("remove(30):", numbers)

# 5. pop() – Remove and return element
popped = numbers.pop()
print("pop():", popped, numbers)

# 6. pop(index)
numbers.pop(1)
print("pop(1):", numbers)

# 7. clear() – Remove all elements
temp = numbers.copy()
temp.clear()
print("clear():", temp)

# 8. index() – Get index of element
print("index(40):", numbers.index(40))

# 9. count() – Count occurrences
numbers.append(40)
print("count(40):", numbers.count(40))

# 10. sort() – Sort list
numbers.sort()
print("sort():", numbers)

# 11. reverse() – Reverse list
numbers.reverse()
print("reverse():", numbers)

# 12. copy() – Shallow copy
copy_list = numbers.copy()
print("copy():", copy_list)

# ===============================
# Extra List Operations
# ===============================

# 13. len()
print("len():", len(numbers))

# 14. min() & max()
print("min():", min(numbers))
print("max():", max(numbers))

# 15. sum()
print("sum():", sum(numbers))

# 16. List slicing
print("slicing [1:4]:", numbers[1:4])

# 17. Membership test
print("40 in list:", 40 in numbers)

# 18. List comprehension
squared = [x * x for x in numbers]
print("list comprehension:", squared)

# ===============================
# End of List Methods Demo
# ===============================
```
