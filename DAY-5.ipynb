## for-loop

```python
name="deep"
for i in name:
    print(i)
```

```python
for i in range(1,6):
    print(i)
```

## while-loop

```python
i=0
while(i<3):
    print(i,end='\t')
    i=i+1   ## i++,i--,--i,++i not in python
```

```python
name=""
while(name!="exit"):
    name=input("enter name:")
    print(name)
else:         ## print("Done")
    print("Done")
```

## break-continue keyword

```python
name = ""
while True:
    name = input("enter name: ")
    if name == "exit":
        break
    print(name)

print("Done")
```

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i,end="\t")
```
