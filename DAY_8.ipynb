# TUPLE

```python
# ===============================
# Python Tuple Demo
# ===============================

# Initial tuple
t = (10, 20, 30, 40, 20)
print("Initial tuple:", t)

# ===============================
# Tuple Methods (Only 2)
# ===============================

# 1. count() – Count occurrences of an element
print("count(20):", t.count(20))

# 2. index() – Get index of first occurrence
print("index(30):", t.index(30))

# ===============================
# Tuple Operations
# ===============================

# 3. len()
print("len():", len(t))

# 4. min() & max()
print("min():", min(t))
print("max():", max(t))

# 5. sum()
print("sum():", sum(t))

# 6. Slicing
print("slicing [1:4]:", t[1:4])

# 7. Concatenation
t2 = (50, 60)
print("concatenation:", t + t2)

# 8. Repetition
print("repetition:", t2 * 3)

# 9. Membership test
print("20 in tuple:", 20 in t)

# ===============================
# Tuple Unpacking
# ===============================
a, b, c, d, e = t
print("unpacking:", a, b, c, d, e)

# ===============================
# Nested Tuple
# ===============================
nested = ((1, 2), (3, 4))
print("nested tuple:", nested)
print("nested access:", nested[0][1])

# ===============================
# Tuple to List Conversion
# ===============================
temp_list = list(t)
temp_list.append(99)
t_modified = tuple(temp_list)
print("tuple after modification:", t_modified)

# ===============================
# Tuple Comprehension (Generator)
# ===============================
gen = (x * x for x in t)
print("generator from tuple:", tuple(gen))

# ===============================
# End of Tuple Demo
# ===============================
```
