## if-elif-else

```python
age=int(input("enter age:").strip(' '))
if(age>18):
    print("allowed")
elif(age>16):
    print("you can do")
else:
    print("no")
    print("NO")
print("Done")
```

## good morning exercise 

```python
import time

# Get current hour (0â€“23)
current_hour = int(time.strftime("%H"))

if current_hour < 12:
    print("Good Morning Sir",time.strftime("%H:%M:%s"))
elif current_hour < 17:
    print("Good Afternoon Sir",time.strftime("%H:%M:%S"))
else:
    print("Good Evening Sir",time.strftime("%H:%M:%S"))
```
