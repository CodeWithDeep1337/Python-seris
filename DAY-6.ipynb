## function argument

### default arguments 

```python
def avarage(a=3,b=5):  
    print((a+b)/2)
    
avarage()
avarage(4)
avarage(b=8)
avarage(10,10)
```

### keyword arguments

```python
def add(a=3,b=5):  
    print(a+b)
add(b=10,a=2) ## given arguments by parameter 
```

### required arguments 

```python
def add(a,b=2):   ## here a is required argument mst given a value.
    print(a+b)
add(1)
```

### variable-length argument 

```python
def Add(*numbers):
    SUM=0
    for i in numbers:
        SUM+=i
    print(SUM)
Add(1,2,3,4)
```
